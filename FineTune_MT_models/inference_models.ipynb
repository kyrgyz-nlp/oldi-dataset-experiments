## load test datasets

```python
source_eng_data = [
    "I went to the cinema with my friends yesterday.",
    "This book turned out to be very useful for me.",
    "The weather today is sunny and warm.",
    "Is there a big garden near your house?",
    "Tomorrow I will go to school in my new school uniform."
]

source_kir_data = ['Мен кечээ досторум менен кино көрүүгө баргам.',
        'Бул китеп мен үчүн абдан пайдалуу болду.',
        'Аба ырайы бүгүн күн ачык жана жылуу.',
        'Сенин үйүңдүн жанында чоң бак барбы?',
        'Мен эртең мектепке жаңы мектеп формасында барам.']
```

## mT5-base-en-ky

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained('alinatl/mT5-base-en-ky',  subfolder="checkpoints/_epoch_6")
model = AutoModelForSeq2SeqLM.from_pretrained("alinatl/mT5-base-en-ky", subfolder="checkpoints/_epoch_6")

model = model.cuda()
model.eval()
print(':)')
```

```python
import torch
import os
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "max_split_size_mb:128"

LANG_TOKEN_MAPPING = {
    'en': '<en>',
    'ky': '<ky>',
}
tokenizer.all_special_tokens
```

```python
def encode_input_str(text, target_lang, tokenizer, seq_len,
                     lang_token_map=LANG_TOKEN_MAPPING):
  target_lang_token = lang_token_map[target_lang]

  input_ids = tokenizer.encode(
      text = target_lang_token + text,
      return_tensors = 'pt',
      padding = 'max_length',
      truncation = True,
      max_length = seq_len)

  return input_ids[0]
```

```python
from torch.utils.data import DataLoader

max_seq_len = 128
target_lang = 'ky'

for i, source in enumerate(source_eng_data):
    print(source)
    target_lang_token = LANG_TOKEN_MAPPING[target_lang]
    source = target_lang_token + source

    token_ids = tokenizer(
        source,
        return_tensors='pt',
        padding='max_length',
        truncation=True,
        max_length=max_seq_len
    )

    with torch.no_grad():
        generated_ids = model.generate(
            token_ids['input_ids'].to('cuda'),
            max_length=max_seq_len
        )
        decoded_preds = tokenizer.batch_decode(
            generated_ids,
            skip_special_tokens=True
        )

    print(decoded_preds)
    print()
```

## NLLB-ky-en

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, NllbTokenizer
```

```python
import gc
import torch

def cleanup():
    """ A function to clean memory (mostly to prevent GPU memory leaks) """
    gc.collect()
    torch.cuda.empty_cache()
cleanup()
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained('alinatl/nllb-ky-en')
tokenizer = NllbTokenizer.from_pretrained('alinatl/nllb-ky-en')
model.to('cuda')
print(':)')
```

```python
model.generation_config.decoder_start_token_id = model.config.decoder_start_token_id
model.config.attention_dropout, model.config.dropout
```

```python
print("decoder_start_token_id:", model.config.decoder_start_token_id)
print("bos_token_id:", tokenizer.bos_token_id)
print("forced_bos_token_id:", model.config.forced_bos_token_id)
```

```python
tokenizer.src_lang = 'kir_Cyrl'
tokenizer.tgt_lang = 'eng_Latn'
tokenizer.src_lang, tokenizer.tgt_lang, model.config.eos_token_id, tokenizer.eos_token, tokenizer.eos_token_id, tokenizer.convert_ids_to_tokens(tokenizer.eos_token_id)
```

```python
max_seq_len = 128

for i, source in enumerate(source_kir_data):
    print(source)
    token_ids = tokenizer(
        source,
        return_tensors='pt',
        padding='max_length',
        truncation=True,
        max_length=max_seq_len
    )['input_ids']

    with torch.no_grad():
        generated_ids = model.generate(input_ids=token_ids.to('cuda'),
                    forced_bos_token_id=tokenizer.convert_tokens_to_ids("eng_Latn"),
                    eos_token_id=tokenizer.eos_token_id,
                  pad_token_id=tokenizer.pad_token_id,
                    max_length=max_seq_len)
        decoded_preds = tokenizer.batch_decode(
            generated_ids,
            skip_special_tokens=True
        )
    print(decoded_preds)
    print()
```

## M2M100-en-ky

```python
from transformers.models.m2m_100.tokenization_m2m_100 import FAIRSEQ_LANGUAGE_CODES

FAIRSEQ_LANGUAGE_CODES["m2m100"] = FAIRSEQ_LANGUAGE_CODES["m2m100"] + ["ky"]
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained('alinatl/m2m100-en-ky')
model = AutoModelForSeq2SeqLM.from_pretrained('alinatl/m2m100-en-ky')
len(tokenizer)
```

```python
new_lang_code = '__ky__'
# Verify the new language has been added
print(f"New language '__ky__' ID: {tokenizer.convert_tokens_to_ids('__ky__')}")
print(f"'__ky__' in additional_special_tokens: {'__ky__' in tokenizer.additional_special_tokens}")
if hasattr(tokenizer, 'lang_code_to_id'):
    print(f"'{new_lang_code}' in lang_code_to_id: {new_lang_code in tokenizer.lang_code_to_id}")
else:
    print("Note: This tokenizer doesn't have a 'lang_code_to_id' attribute.")
```

```python
new_lang_id = tokenizer.convert_tokens_to_ids('__ky__')
tokenizer.lang_code_to_id[new_lang_code] = new_lang_id
tokenizer.lang_code_to_token["ky"] = "__ky__"
tokenizer.lang_token_to_id["__ky__"] = tokenizer.convert_tokens_to_ids("__ky__")
```

```python
model.generation_config
```

```python
tokenizer.src_lang = 'en'
tokenizer.tgt_lang = 'ky'
model.to('cuda')
model.eval()
```

```python
max_seq_len = 128

for i, source in enumerate(source_eng_data):

    print(source)
    token_ids = tokenizer(
        source,
        return_tensors='pt',
        padding='max_length',
        truncation=True,
        max_length=max_seq_len
    )['input_ids']

    with torch.no_grad():
        generated_ids = model.generate(
            token_ids.to('cuda'),
            forced_bos_token_id=tokenizer.get_lang_id("ky"),
            max_length=max_seq_len
        )
        decoded_preds = tokenizer.batch_decode(
            generated_ids,
            skip_special_tokens=True
        )

    print(decoded_preds)
    print()
```
